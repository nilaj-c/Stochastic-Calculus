# Wiener Process:

Here we would code our first example of a Stochastic process and the simplest, a Wiener process. In the terminology of Stochastic Calculus, a Wiener process is a one-dimensional Brownian motion. It has the following properties:

1. $W(0) = 0$

2. $W_t$ is continuous for all t

3. For $ t \geq 0$ and $\Delta t \geq 0$, the increment $W(t + \Delta t) - W(t)$ is normally distributed with mean 0 and variance $\Delta t$

4. The increments $W(t_n) - W(t_{n-1})$ are independently distributed

We will simulate some sample paths of a Wiener process in the following section.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Defining the Wiener process function:

```python
def wiener_process(end_time, n_steps):
    
    #inputs: end_time: Total time
    #        n_steps: number of time steps
    
    #output: W : a list containing the Wiener process path
    
    W0 = [0]
    
    dt = end_time/n_steps
    
    increments = np.random.normal(0.0, 1.0*np.sqrt(dt), n_steps) #simulate a normally distributed list of increments
    
    W = W0 + list(np.cumsum(increments))
    
    time = np.linspace(0.0,end_time,num=n_steps+1,endpoint=True) 
    
    return (time,W)
    
```

#### Plotting some sample paths



```python
t_end = 10.0
n_steps = 10000

plt.figure(figsize=(12,8))

for _ in range(5):
    plt.plot(wiener_process(t_end,n_steps)[0], wiener_process(t_end,n_steps)[1])
    
plt.xlabel('Time (t)', size = 18)
plt.ylabel('W(t)', size = 18)
plt.tick_params(labelsize=18)
```
